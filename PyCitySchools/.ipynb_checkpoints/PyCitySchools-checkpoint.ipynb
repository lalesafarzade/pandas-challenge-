## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

```python
Total_Students=len(student_data['Student ID'])
#making a dictionary of indexes and the values
summary_dict= {'Total Schools': [len(school_data['school_name'].unique())],
               'Total Students': "{:,}".format(Total_Students),
               'Total Budget'  : "${:,}.00".format(sum(school_data['budget'])),
                'Average Math Score': round(sum(student_data['math_score'])/Total_Students,6),
                'Average Reading Score': round(sum(student_data['reading_score'])/Total_Students,6),
                '% Passing Math': round(len(student_data[student_data['math_score']>=70])/Total_Students,8)*100,
                '% Passing Reading': round(len(student_data[student_data['reading_score']>=70])/Total_Students,8)*100,
                '% Overall Passing': round(len(student_data[(student_data['reading_score']>=70)&(student_data['math_score']>=70)])/Total_Students,8)*100
                         
                         
                         }
df=pd.DataFrame(summary_dict)
df
```

## School Summary

* Creating an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Creating a dataframe to hold the above results

```python
#making a type_dataframe from school_data 
df_1=school_data[['school_name','type']].set_index('school_name').sort_index().rename(columns={"type": "School Type"}) 

#making a dataframe from school_data using groupby()
df_2=school_data.groupby ('school_name').sum().drop(['School ID'],axis=1).rename(columns=\
                  {"budget": "Total School Budget", "size": "Total Students"})

df_2['Per Student Budget']= df_2['Total School Budget']/df_2['Total Students']

#making a dataframe from student_data using groupby()
df_3=student_data.groupby('school_name').mean().drop(['Student ID'],axis=1).rename(columns=\
                    {"reading_score": "Average Reading Score", "math_score": "Average Math Score"})

#merging the 3 dataframes
df_4=functools.reduce(lambda  left,right: pd.merge(left,right,left_index=True,right_index=True), [df_1,df_2,df_3])

#adding new columns using groupby()
df_4['% Passing Math']=(student_data[student_data['math_score']>=70].groupby('school_name').count()['math_score']/\
                        df_4['Total Students'])*100

df_4['% Passing Reading']=(student_data[student_data['reading_score']>=70].groupby('school_name').count()['reading_score']/\
                            df_4['Total Students'])*100

df_4['% Overall Passing']=(student_data[(student_data['reading_score']>=70)&(student_data['math_score']>=70)]\
                           .groupby('school_name').count()['reading_score']/df_4['Total Students'] )*100

school_summary =df_4.reindex(columns=['School Type','Total Students','Total School Budget','Per Student Budget',
                                      'Average Math Score','Average Reading Score',  '% Passing Math',
                                                     '% Passing Reading', '% Overall Passing'])
#formatting some columns
school_summary_display=school_summary.copy()
school_summary_display['Total School Budget']=["${:,}.00".format(i) for i in school_summary_display['Total School Budget']]
school_summary_display['Per Student Budget']=["${:.2f}".format(i) for i in school_summary_display['Per Student Budget']]
school_summary_display
```

## Top Performing Schools (By % Overall Passing)

## Math Scores by Grade

* Create a table that lists the average Math Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
#FirstWay: using hint

data_frames=[student_data[student_data['grade']==i][['math_score','school_name']].groupby('school_name').mean() for i in ['9th','10th','11th','12th']]
Math_score_by_grade = functools.reduce(lambda  left,right: pd.merge(left,right,left_index=True,right_index=True), data_frames)
Math_score_by_grade.columns=['9th','10th','11th','12th']
Math_score_by_grade
```

## Reading Score by Grade 

* Perform the same operations as above for reading scores

```python
#second way:using_groupby
import numpy as np
Data_series=student_data.groupby(['grade','school_name']).mean()['reading_score']
data=list(Data_series)
r=list(Data_series.index)
index_=list({i[0]:r.count(i[0]) for i in r}.keys())
columns_=list({i[1]:r.count(i[1]) for i in r}.keys())
u=pd.DataFrame(np.reshape(data,(4,15)),index=index_,columns=columns_).T
#reorganising the columns
Reading_score_by_grade=u.reindex(['9th', '10th', '11th', '12th'],axis='columns')
Reading_score_by_grade
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
bins = [0 , 584, 629, 644, 680]
# Create the names for the four bins
group_names = ["<$585", "$585-630", "$630-645", '$645-680']
school_Spending=school_summary.copy()
school_Spending['Spending Ranges (Per Student)']=pd.cut(school_Spending["Per Student Budget"], bins, labels=group_names,include_lowest=True)
Scores_By_School_Spending=school_Spending.groupby('Spending Ranges (Per Student)').mean()[['Average Math Score','Average Reading Score','% Passing Math','% Passing Reading','% Overall Passing']]
Scores_By_School_Spending.round(2)
```

## Scores by School Size

* Perform the same operations as above, based on school size.

```python
bins = [0 , 999, 1999, 5000]
# Create the names for the three bins
group_names = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
school_size=school_summary.copy()
school_size['School Size']=pd.cut(school_size["Total Students"], bins, labels=group_names,include_lowest=True)
Scores_By_School_Size=school_size.groupby('School Size').mean()[['Average Math Score','Average Reading Score','% Passing Math','% Passing Reading','% Overall Passing']]
Scores_By_School_Size
```

## Scores by School Type

* Perform the same operations as above, based on school type

```python
school_type=school_summary.copy()
Scores_by_School_Type=school_type.groupby('School Type').mean()[['Average Math Score','Average Reading Score','% Passing Math','% Passing Reading','% Overall Passing']]
Scores_by_School_Type
```
